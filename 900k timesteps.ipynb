```python
import gymnasium as gym

env = gym.make("CarRacing-v3", continuous=True)
obs, info = env.reset()
print("CarRacing loaded!")
```

```python
from stable_baselines3 import PPO
env = gym.make("CarRacing-v3", continuous=True)

model = PPO.load("ppo_carracing_800k.zip", env=env, device="cuda")
print(model.num_timesteps)
```

```python
model.lr_schedule = lambda _: 1e-4
print("Learning rate set to 0.0001")
```

```python
remaining = 1_000_000 - model.num_timesteps
print("Remaining steps:", remaining)

model.learn(
    total_timesteps=remaining,
    reset_num_timesteps=False
)
```

```python
model.save("ppo_carracing_900k.zip")
print("Saved 900k!")
```
